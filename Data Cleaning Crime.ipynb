This notebook contains the code for cleaning the crime data from the FBI database. The raw data is available as a csv/excel file from the FBI website. 

```python
#Importing Libraries
import pandas as pd
import numpy as np
import csv as csv
import matplotlib.pyplot as plt
```

```python
#Reading the raw data
df = pd.read_excel('/Users/bhargavsrisainama/Documents/personal/CU Boulder Data Science/STAT/Project/Crime/State wise/Copy of 05tbl05.xls')
```

```python
df.head()
```

```python
#Returing the states names
df["State"]
```

```python
#now this has merged elements with NaN values in State column , replace it with the corresponding state name 
df['State'].fillna(method='ffill', inplace=True)
```

```python
df["Area"].fillna(method='ffill', inplace=True) 
```

```python
#This is to ensure the data is only put up in the required column
df["Area"] == "State Total"
```

```python
#drop the rows where the area is not "State Total" 
df = df[df.Area == "State Total"]
```

```python
df
```

```python
#drop the coumns in Popuulation which are NaN
df = df[df.Population.notnull()]
```
