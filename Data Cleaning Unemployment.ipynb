```python
df.drop(df.index[0:5],inplace = True)
```

```python
#Returning the last 2 elements of the string in the column "County Name" to get the state code
df["County Name"] = df["County Name"].str[-2:]
```

```python
#remove the rows where the county name is "PR"
df = df[df["County Name"] != "PR"]
```

```python
#Now we are grouping the data by the county name and summing the values of the labor force, employed, and unemployed columns
state_totals_complete = df.groupby('County Name')[['Labor Force', 'Employed', 'Unemployed']].sum().reset_index()

# Display the resulting dataframe with totals for each state (as represented by 'County Name')
print(state_totals_complete)
```
